```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
Data1 = np.random.randn(n, 2)
Data2 = np.random.randn(n, 2) + 5
```

```python
Label1 = np.zeros((n, 1), dtype=np.int)
Label2 = np.ones((n, 1), dtype=np.int)
```

```python
TrainData = np.vstack([Data1, Data2]) 
LabelData = np.vstack([Label1, Label2]) 
print(np.size(TrainData, 0))
```

```python
TrainData = np.vstack([Data1, Data2]) # 2000 , 2
LabelData = np.vstack([Label1, Label2]) # 2000 , 1

num_data = np.size(TrainData, 0) # 2000

plt.plot(Data1[:, 0], Data1[:, 1], "r*")
plt.plot(Data2[:, 0], Data2[:, 1], "g*")

feature_num = 2
num_output = 1
w = np.random.randn(feature_num, num_output)
b = np.random.randn(1)

print(w, b)

x1 = np.arange(-3, 8) # -3 ~ 7
x2 = (-w[0] * x1 - b) / w[1]

plt.plot(x1, x2)
# 선 이름은 초평면, 결정 경졔
```

```python
[i for i in [w, b, x1, x2]]
```

```python
def step(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
for i in range(num_data):
    x = TrainData[i] 
    target = LabelData[i]
    print(x)
    print(target)
```

```python
num_epoch = 100
learning_rate = 0.5
for epoch in range(num_epoch):
    for i in range(num_data):
        x = TrainData[i] # (2, 1) 
        target = LabelData[i] # 0 or 1
        
        y = np.dot(x, w) + b # (2, 1) 내적 (2, 1) wx + b(=ax + b)
        z = step(y)
        e = target - z
        
        delta_w = learning_rate * e * np.transpose([x]) 
        delta_b = learning_rate * e * 1
        
        w = w + delta_w
        b = b + delta_b
```

```python
plt.plot(Data1[:, 0], Data1[:, 1], "r*")
plt.plot(Data2[:, 0], Data2[:, 1], "g*")

x1 = np.arange(-3, 8)
x2 = (-w[0]*x1 - b) / w[1]
plt.plot(x1, x2, 'r')
```
